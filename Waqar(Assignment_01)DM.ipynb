Aim

To apply the K-Means clustering algorithm using the scikit-learn library on a given dataset and analyze the results.

Dataset

| Point | Coordinates |
| ----- | ----------- |
| A     | (2,10)      |
| B     | (2,5)       |
| C     | (8,4)       |
| D     | (5,8)       |
| E     | (7,5)       |
| F     | (16,-4)     |

Number of clusters: K = 3

Initial centroids:

A (2,10)
B (2,5)
D (5,8)

Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Create Dataset

```python
X = np.array([
    [2,10],   # A
    [2,5],    # B
    [8,4],    # C
    [5,8],    # D
    [7,5],    # E
    [16,-4]   # F
])

labels = ['A','B','C','D','E','F']
```

Apply K-Means (K = 3)

```python
init_centroids = np.array([
    [2,10],   # A
    [2,5],    # B
    [5,8]     # D
])

kmeans = KMeans(
    n_clusters=3,
    init=init_centroids,
    n_init=1,
    random_state=0
)

kmeans.fit(X)
```

Display Results

```python
print("Cluster Labels:")
for i in range(len(X)):
    print(labels[i], "-> Cluster", kmeans.labels_[i])

print("\nFinal Centroids:")
print(kmeans.cluster_centers_)

print("\nSum of Squared Errors (SSE):")
print(kmeans.inertia_)
```

Visualization

```python
plt.figure()

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
for i, txt in enumerate(labels):
    plt.annotate(txt, (X[i][0], X[i][1]))

centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], marker='X', s=200)

plt.title("K-Means Clustering")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.spatial.distance import pdist, squareform
```

Step 1: Define Data Points

```python
labels = ['A', 'B', 'C', 'D', 'E', 'F']

X = np.array([
    [7, 8],  # A
    [3, 5],  # B
    [6, 4],  # C
    [6, 0],  # D
    [1, 5],  # E
    [8, 4]   # F
])
```

Step 2: Compute Distance Matrix

```python
dist_matrix = pdist(X, metric='euclidean')
print("Pairwise Distance Matrix:\n")
print(squareform(dist_matrix))
```

Single Linkage

```python
plt.figure(figsize=(8, 5))
single_link = linkage(X, method='single', metric='euclidean')
dendrogram(single_link, labels=labels)
plt.title("Dendrogram - Single Linkage")
plt.xlabel("Points")
plt.ylabel("Distance")
plt.show()
```

Average Linkage

```python
plt.figure(figsize=(8, 5))
average_link = linkage(X, method='average', metric='euclidean')
dendrogram(average_link, labels=labels)
plt.title("Dendrogram - Average Linkage")
plt.xlabel("Points")
plt.ylabel("Distance")
plt.show()
```

Centroid Linkage

```python
plt.figure(figsize=(8, 5))
centroid_link = linkage(X, method='centroid', metric='euclidean')
dendrogram(centroid_link, labels=labels)
plt.title("Dendrogram - Centroid Linkage")
plt.xlabel("Points")
plt.ylabel("Distance")
plt.show()
```

Are the dendrograms different?

```python
print("""
    YES
    
Reason:
Single link uses minimum distance
→ causes chaining effect
→ clusters join earlier

Average link uses average distance
→ produces compact clusters

Centroid link uses center of clusters
→ cluster shape influences merging""")
```
